```python
#Independent and dependent feaure split
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
#Train test split-->
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.head()
```

# Standard scalling-->

```python
#Standard scalling-->
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
import pickle
pickle.dump(scaler,open("scaling.pkl","wb"))
```

```python
x_train
```

```python
x_test
```

# Model training

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# print the coefficients the intercepts
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
#On which parameters the model has been train
model.get_params()
```

# predection with test data

```python
reg_predict=model.predict(x_test)
```

```python
reg_predict
```

```python
#Plot a scatter plot for analyize predection
plt.scatter(y_test,reg_predict)
```

```python
#Predection with residuals
residuals=y_test-reg_predict
```

```python
residuals
```

```python
#ploting this residual
sns.displot(residuals,kind="kde")
```

```python
#scatter plot with respect to prediction vs residual 
#uniform distribution
plt.scatter(reg_predict,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,reg_predict))
print(mean_squared_error(y_test,reg_predict))
print(np.sqrt(mean_squared_error(y_test,reg_predict)))
```

# Adjusted r square and r square
formula -->
R^2 = 1 -SSR/SST
R^2-coefficient of determination
SSR- sum of square of residuals
SST- Sum of square total
Adjusted R2 = 1 - [(1-R2)*(n-1)/(n-k-1)]

```python
from sklearn.metrics import r2_score
score= r2_score(y_test,reg_predict)
print(score)
```
